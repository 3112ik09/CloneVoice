```python
import pandas as pd
import numpy as np
```

```python
import nltk
from nltk.corpus import indian
indian_languages = indian.fileids()
hindi_words = indian.words('hindi.pos')
```

```python
def hiEng(hindi):
    output = []
    for word in hindi.split(" "):
        if word in hindi_words:
            output.append(word)
        else:
            translation = translator.translate(word, src="hi", dest="en")
            english_word = translation.text
            output.append(english_word)
#     print(output)
    HiEng = " ".join(output)
    print(HiEng)
hiEng(hindi)
    
```
